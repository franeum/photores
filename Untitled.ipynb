```python
import pandas as pd
```

```python
path = './log.txt'
path_long = './log_11mt_avg16.txt'
path_short = './log_1mt_avg16.txt'
```

```python
with open(path_short,'r',encoding='utf-8') as f:
    short = f.readlines()
    
with open(path_long,'r',encoding='utf-8') as f:
    long = f.readlines()
```

```python
values_short = []
values_long = []

for i in short:
    d = i.replace('),\n','').split(' ')[-1]
    values_short.append(float(d))
    
for i in long:
    d = i.replace('),\n','').split(' ')[-1]
    values_long.append(float(d))
```

```python
len(values_short) == len(values_long)
```

```python
len(values_short),len(values_long)
```

```python
with open(path,'r',encoding='utf-8') as f:
    data = f.readlines()
    
values = []
for i in data:
    t = i.split(" ")[1].replace("'","").replace(",","")
    t = datetime.strptime(t, "%H:%M:%S").time()
    val = i.replace('),\n','').split(' ')[-1]
    values.append([t,float(val)])
```

```python
df2 = pd.DataFrame(values)
```
